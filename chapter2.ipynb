```python
import os
import numpy as np
import matplotlib.pyplot as plt

from utils import normal
```

```python
xs = np.loadtxt('data/height.txt')
print(xs.shape)
```

```python
plt.hist(xs, bins='auto', density=True)
plt.xlabel('Height(cm)')
plt.ylabel('Probability Density')
plt.show()
```

```python
mu = np.mean(xs)
sigma = np.std(xs)

x = np.linspace(150, 190, 1000)
y = normal(x, mu=mu, sigma=sigma)
samples = np.random.normal(loc=mu, scale=sigma, size=10000)
print(mu, sigma)
```

```python
plt.hist(xs, bins='auto', density=True, alpha=0.3, label='original')
plt.hist(samples, bins='auto', density=True, alpha=0.3, label='generated', color='r')
plt.plot(x, y)
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.show()
```
