```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(np.array([1,2,3]))
```

```python
def normal(x, mu=0, sigma=1):
    y = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))
    return y
```

```python
x = np.linspace(-5, 5, 100)
y0 = normal(x, mu=0, sigma=1)
y1 = normal(x, mu=0, sigma=0.5)
y2 = normal(x, mu=0, sigma=2)

```

```python
plt.plot(x, y0, label='$\sigma$=1')
plt.plot(x, y1, label='$\sigma$=0.5')
plt.plot(x, y2, label='$\sigma$=2')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x_means = []
N = 10

for _ in range(10000):
    xs = []
    for n in range(N):
        x = np.random.rand()
        xs.append(x)
    mean = np.mean(xs)
    x_means.append(mean)
```

```python
plt.hist(x_means, bins='auto', density=True)
plt.xlim(-0.05, 1.05)
plt.ylim(0, 5)
plt.show()
```

```python
x_sums = []
N = 5

for _ in range(10000):
    xs = []
    for n in range(N):
        x = np.random.rand()
        xs.append(x)
    t = np.sum(xs)
    x_sums.append(t)
```

```python
x_norm = np.linspace(-5, 5, 1000)
mu = N / 2
sigma = np.sqrt(N / 12)
y_norm = normal(x_norm, mu, sigma)
```

```python
plt.hist(x_sums, bins='auto', density=True)
plt.plot(x_norm, y_norm)
plt.xlim(-1, 6)
plt.show()
```
